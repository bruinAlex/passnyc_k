```python
# Basic packages
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import random as rd # generating random numbers
import datetime # manipulating date formats

# Viz
import matplotlib.pyplot as plt # basic plotting
import seaborn as sns # for prettier plots

# Multiple outputs from single cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# Import both datasets.
schoolExplorerDat = pd.read_csv(r'data/2016 School Explorer.csv')
schoolExplorerDat.head()
schoolExplorerDat.shape
```

## School Explorer
1,272 schools with 161 features each. Yikes.

```python
# Allow pandas to show up to 161 columns and rows without truncating
pd.set_option('max_columns', 161)
pd.set_option('max_rows', 161)
```

```python
schoolExplorerDat.info()
schoolExplorerDat.describe()
```

```python
schoolExplorerDat.dtypes
```

```python
schoolExplorerEcon = schoolExplorerDat.loc[:, ['Economic Need Index', 'School Income Estimate', 'Percent Black / Hispanic']]
schoolExplorerEcon['School Income Estimate'] = schoolExplorerEcon['School Income Estimate'].replace('[\$,]', '', regex=True).astype(float) # Convert currency to floats
schoolExplorerEcon['Percent Black / Hispanic'] = schoolExplorerEcon['Percent Black / Hispanic'].apply(lambda x: float(x.strip('%'))/100)
```

```python
schoolExplorerEconNoNAN = schoolExplorerEcon.dropna()

```

```python
sns.pairplot(schoolExplorerEconNoNAN,
            height = 5,
            dropna=True,
            vars=['Economic Need Index', 'School Income Estimate','Percent Black / Hispanic'])
```
